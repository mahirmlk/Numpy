## operations with business example

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Data structure: [restaurant_id, 2021, 2022, 2023, 2024]"

sales_data = np.array([
    [1, 150000, 180000, 220000, 250000],  # Paradise Biryani\n",
    [2, 120000, 140000, 160000, 190000],  # Beijing Bites\n",
    [3, 200000, 230000, 260000, 300000],  # Pizza Hub\n",
     [4, 180000, 210000, 240000, 270000],  # Burger Point\n",
    [5, 160000, 185000, 205000, 230000]   # Chai Point\n",
    ])

print("----Sales Data Analysis----")    
print("\n Sales Data Shape:\n", sales_data.shape)
print("\n Sample data for first 3 restaurants:\n", sales_data[:3])
print("\n Sales Amount: \n", sales_data[:, 1: ])

# Calculate total sales for each year
total_sales_per_year = np.sum(sales_data[:, 1:], axis=0)
years = [2021, 2022, 2023, 2024]
print("\n Total Sales Per Year:\n", total_sales_per_year)

# minimmum sales per restaurant
min_sales_per_restaurant = np.min(sales_data[:, 1:], axis=1)
print("\n Minimum Sales Per Restaurant:\n", min_sales_per_restaurant)

# maximum sales per year
max_sales_per_year = np.max(sales_data[:, 1:], axis=0)
print("\n Maximum Sales Per Year:\n", max_sales_per_year)

# Average sales per restaurant
avg_sales_per_restaurant = np.mean(sales_data[:, 1:], axis=1)
print("\n Average Sales Per Restaurant:\n", avg_sales_per_restaurant)

# cumulative sales per restaurant
cumulative_sales = np.cumsum(sales_data[:, 1:], axis=1)
print("\n Cumulative Sales Per Restaurant:\n", cumulative_sales)
```

### matplotlib plotting

```python
plt.figure(figsize=(10, 6))
plt.plot(years, total_sales_per_year, marker='o'
, label='Total Sales Per Year')
plt.title('Total Sales Per Year (2021-2024)')
plt.xlabel('Year')
plt.ylabel('Total Sales Amount')
plt.xticks(years)
plt.grid()
plt.legend()
plt.show()
```

```python
plt.figure(figsize= (10, 6))
plt.plot(np.mean(cumulative_sales, axis=0))
plt.title('Average Cumulative Sales Per Year')
plt.xlabel('Years')
plt.ylabel('Average Cumulative Sales')
plt.xticks(ticks= range(4), labels= years)
plt.grid()
plt.show()

```

### Vectors

```python
vector1 = np.array([1, 2, 3, 4, 5])
vector2 = np.array([6, 7, 8, 9, 10])

print("Vector Addition:", vector1 + vector2)
print("Vector Subtraction:", vector2 - vector1)
print(" Multiplication:", vector1 * vector2)

print(" Dot Product:", np.dot(vector1, vector2))
print(" Magnitude of Vector1:", np.linalg.norm(vector1))
print(" Magnitude of Vector2:", np.linalg.norm(vector2))
print(" Angle between Vector1 and Vector2:", 
      np.arccos(np.dot(vector1, vector2) / (np.linalg.norm(vector1) * np.linalg.norm(vector2))))
```
