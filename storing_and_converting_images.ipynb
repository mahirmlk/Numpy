```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
array1 = np.array([[1, 2, 3], [4, 5, 6]])
array2 = np.random.rand(3, 3)
array3 = np.zeros((4, 4))
```

```python
np.save('array1.npy', array1)
np.save('array2.npy', array2)
np.save('array3.npy', array3)
```

```python
loaded_array1 = np.load('array1.npy')
loaded_array2 = np.load('array2.npy')
loaded_array3 = np.load('array3.npy')
print("Loaded Array 1:\n", loaded_array1)
print("Loaded Array 2:\n", loaded_array2)
print("Loaded Array 3:\n", loaded_array3)
```

```python
try:
    logo = np.load('numpy-logo.npy')

    # Display the image
    plt.figure(figsize=(6, 6))
    plt.subplot(121)
    plt.title("Numpy Logo from Numpy Array")
    plt.imshow(logo)
    plt.grid(False)
    plt.axis('off')

    dark_logo = 1 - logo
    plt.subplot(122)
    plt.title("Inverted Numpy Logo")
    plt.imshow(dark_logo)
    plt.grid(False)
    plt.axis('off')


except FileNotFoundError:
    print("Numpy logo file not found.")
```
